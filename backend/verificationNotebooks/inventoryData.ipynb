```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\gaura\Desktop\ai-agent\backend\inventory_data.csv")
```

```python
df.head()
```

## 1) Which product has the lowest stock level?

```python
df.groupby("Product")["Stock_Level"].sum().idxmin()
```

## 2) Which products are below reorder point?

```python
df[df["Stock_Level"] < df["Reorder_Point"]].Product
```

## 3) Show warehouse cost by region as pie chart

```python
# Show warehouse cost by region as pie chart

warehouse_cost = df.groupby("Region")["Warehouse_Cost"].sum()

plt.pie(
    warehouse_cost,
    labels=warehouse_cost.index,
    autopct='%1.1f%%'
)

plt.title("Warehouse Cost by Region")

plt.show()
```

## 4) What is the stock trend over time?

```python
df["Date"] = pd.to_datetime(df["Date"])

stock_trend = df.groupby("Date")["Stock_Level"].sum()

plt.plot(
    stock_trend.index,
    stock_trend.values
)

plt.title("Stock Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Level")

plt.show()
```
