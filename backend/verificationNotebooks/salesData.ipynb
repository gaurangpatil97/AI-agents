```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\gaura\Desktop\ai-agent\backend\sales_data.csv")
```

```python
df.head()
```

## 1) Which product had highest sales?


```python
print(df.groupby("Product")['Total Sales'].sum().sort_values(ascending=False).iloc[0])
```

## 2) Show me sales by region as pie chart


```python
region_sales = df.groupby("Region")["Total Sales"].sum()

plt.pie(
    region_sales,
    labels=region_sales.index,
    autopct='%1.1f%%'
)

plt.show()
```

## 3) What is the monthly sales trend of the year 2023?

```python
df["Date"] = pd.to_datetime(df["Date"])

sales_2023 = df[df["Date"].dt.year == 2023]

monthly_sales = sales_2023.groupby(
    sales_2023["Date"].dt.month
)["Total Sales"].sum()

plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker='o'
)

plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend - 2023")

plt.show()
```

## 4) Which age group spends the most?

```python
age_sales = df.groupby("Age Group")["Total Sales"].sum()

print(age_sales.sort_values(ascending=False).iloc[0])
```
